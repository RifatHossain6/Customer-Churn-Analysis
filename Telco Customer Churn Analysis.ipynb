```python
# Import essential Python libraries for data analysis and visualization
import pandas as pd       # For data manipulation and analysis
import numpy as np        # For numerical computations
import matplotlib.pyplot as plt  # For creating basic visualizations
import seaborn as sns     # For advanced statistical data visualizations
```

```python
# Load the CSV file 'Customer Churn.csv' into a pandas DataFrame named df
df =pd.read_csv('/content/Customer Churn.csv')
```

```python
# Replace any blank spaces (" ") in 'TotalCharges' column with 0
# Then convert the 'TotalCharges' column data type to float for numerical operations
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype(float)
```

```python
# This line checks for duplicate customer IDs and counts how many times they appear after the first occurrence
df["customerID"].duplicated().sum()
```

```python
# Convert the 'SeniorCitizen' column values: 1 becomes 'Yes', 0 becomes 'No'
def conv(value):
  if value==1:
    return "Yes"
  else:
      return "No"
df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
# Create a count plot to show the number of customers for each 'Churn' category (Yes/No)
ax = sns.countplot(x='Churn', data=df)

# Add numeric labels on top of each bar to show exact counts
ax.bar_label(ax.containers[0])

# Set the title of the plot
plt.title("Count of customers by churn")

# Display the plot
plt.show()
```

```python
# Create a pie chart to show the percentage of churned vs non-churned customers
plt.figure(figsize=(3, 4))  # Set the figure size

# Group the data by 'Churn' and count the number of entries in each group
gb = df.groupby('Churn').agg({'Churn': "count"})

# Plot the pie chart using the grouped churn counts
plt.pie(gb['Churn'], labels=gb.index, autopct="%1.2f%%")  # Show percentages with 2 decimal places

# Add a title to the chart
plt.title("Percentage of churned customers", fontsize=10)

# Display the pie chart
plt.show()
```

```python
# Create a count plot to show the number of customers by gender
plt.figure(figsize=(5, 5))  # Set the size of the plot

ax = sns.countplot(x="gender", data=df)  # Create the countplot for gender

ax.bar_label(ax.containers[0])  # Add count labels on top of each bar

plt.title("Count of Customers by Gender")  # Add a title to the plot

plt.show()  # Display the plot
```

```python
# Create a count plot to compare churn status between senior and non-senior citizens
plt.figure(figsize=(3.5, 4))  # Set the plot size

ax = sns.countplot(x="SeniorCitizen", data=df, hue="Churn")  # Create a grouped bar chart by churn

# Add count labels on top of each bar for both churn categories
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

# Add a title to the chart
plt.title("Churn by SeniorCitizen")

# Display the plot
plt.show()
```

```python
# Step 1: Crosstab counts
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"])

# Step 2: Row-wise percentage
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Step 3: Plot
ax = ct_percent.plot(kind='bar', stacked=True, figsize=(4, 4), )

# Step 4: Add percentage labels (corrected index use)
for i, idx in enumerate(ct.index):  # `idx` is the actual index like 0 or 1
    bottom = 0
    for col in ct.columns:
        value = ct_percent.loc[idx, col]
        label = f"{value:.1f}%"
        height = value
        plt.text(i, bottom + height / 2, label, ha='center', va='center', fontsize=8)
        bottom += height

plt.title("Churn by SeniorCitizen (Stacked Bar Chart)")
plt.ylabel("Percentage")
plt.xlabel("SeniorCitizen")
plt.legend(title="Churn",  bbox_to_anchor=(0.9, 0.9),)
plt.tight_layout()
plt.show()
```

```python
# Create a histogram to show the distribution of customer tenure, separated by churn status
plt.figure(figsize=(9, 4))  # Set the size of the plot

# Plot the histogram of 'tenure' with 72 bins, colored by 'Churn'
sns.histplot(x="tenure", data=df, bins=72, hue="Churn")

plt.title("Distribution of Tenure")  # Add a title to the plot

plt.show()  # Display the plot
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# subplot layout: 3 rows x 3 columns = 9 total
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))

# axes is a 2D array, flatten it for easy looping
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(x=col, hue="Churn", data=df, ax=axes[i])
    axes[i].set_title(f"Churn by {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=20)

# Adjust layout
plt.tight_layout()
plt.show()
```
